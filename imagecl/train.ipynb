```python
from tkinter import Button
import gradio as gr

def classify_image(results):
    labels_dict = results[0].names
    probs = results[0].probs.data.tolist()
    detected_objects = []

    for i in range(len(probs)):
        if probs[i]>0.80:
            detected_objects.append(results[0].names[i])
    return {"landmarks":list(detected_objects)}
iface = gr.Interface(
    fn=classify_image,
    inputs=gr.Image(label="Query image"),
    outputs="text",
    title="Image Classification for chocolate raper",
    description="Upload an image for classification.",
    theme="compact"
)

# Launch the interface
iface.launch()
```
